```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# Assuming you have a dataset stored in a CSV file
dataset_path = 'career_pred.csv'

# Load the dataset into a pandas DataFrame
df = pd.read_csv(dataset_path)

# Extract the column data
column_name = 'sslc'  # Replace 'column_name' with the actual column name in your dataset
column_data = df[column_name]

# Calculate the unique values
values = np.unique(column_data)


# Calculate statistics
mean_value = np.mean(column_data)
mode_value = column_data.mode().values[0]
std_value = np.std(column_data)

# Create an array to store the heights of the bars
bar_heights = column_data.values

# Create the bar chart with custom heights
plt.bar(range(len(column_data)), bar_heights)

# Add a horizontal line for the mean value
plt.axhline(mean_value, color='red', linestyle='--', label='Mean')

# Add a vertical line for the mode value
plt.axvline(mode_value, color='green', linestyle='--', label='Mode')

# Set the labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Bar Chart of {}'.format(column_name))
plt.legend()  # Add a legend

# Display the chart
plt.show()

# Print the statistics
print('Mean:', mean_value)
print('Mode:', mode_value)
print('Standard Deviation:', std_value)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn import metrics

# Assuming you have a dataset stored in a CSV file
dataset_path = 'career_pred.csv'

# Load the dataset into a pandas DataFrame
df = pd.read_csv(dataset_path)

# Select the columns you want to use for clustering
columns = ['sslc', 'hsc', 'cgpa']

# Define the range of values of K to test
k_values = range(2, 10)

# Create empty dictionaries to store the evaluation scores
silhouette_scores = {}
calinski_scores = {}
davies_scores = {}

# Iterate over each value of K
for k in k_values:
    # Perform clustering using K-means
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df[columns])
    
    # Compute Silhouette Coefficient
    silhouette_scores[k] = metrics.silhouette_score(df[columns], kmeans.labels_)
    
    # Compute Calinski-Harabasz Index
    calinski_scores[k] = metrics.calinski_harabasz_score(df[columns], kmeans.labels_)
    
    # Compute Davies-Bouldin Index
    davies_scores[k] = metrics.davies_bouldin_score(df[columns], kmeans.labels_)

# Print the evaluation scores
print("Silhouette Coefficient:")
for k, score in silhouette_scores.items():
    print(f"K = {k}: {score:.4f}")
    
print("\nCalinski-Harabasz Index:")
for k, score in calinski_scores.items():
    print(f"K = {k}: {score:.4f}")
    
print("\nDavies-Bouldin Index:")
for k, score in davies_scores.items():
    print(f"K = {k}: {score:.4f}")
```

```python
import matplotlib.pyplot as plt

# Extract the values of K and Davies-Bouldin scores
k_values = list(davies_scores.keys())
davies_scores_values = list(davies_scores.values())

# Plot the elbow graph
plt.plot(k_values, davies_scores_values, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Davies-Bouldin Index')
plt.title('Elbow Graph - Davies-Bouldin Index')
plt.xticks(k_values)
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn import metrics
import matplotlib.pyplot as plt

# Assuming you have a dataset stored in a CSV file
dataset_path = 'career_pred.csv'

# Load the dataset into a pandas DataFrame
df = pd.read_csv(dataset_path)

# Select the columns you want to use for clustering
columns = ['sslc', 'hsc', 'cgpa']

# Define the range of values of K to test
k_values = range(2, 10)

# Create empty dictionaries to store the evaluation scores
silhouette_scores = {}
calinski_scores = {}
davies_scores = {}
wcss = []

# Iterate over each value of K
for k in k_values:
    # Perform clustering using K-means
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df[columns])
    
    # Compute Silhouette Coefficient
    silhouette_scores[k] = metrics.silhouette_score(df[columns], kmeans.labels_)
    
    # Compute Calinski-Harabasz Index
    calinski_scores[k] = metrics.calinski_harabasz_score(df[columns], kmeans.labels_)
    
    # Compute Davies-Bouldin Index
    davies_scores[k] = metrics.davies_bouldin_score(df[columns], kmeans.labels_)
    
    # Compute the WCSS (inertia)
    wcss.append(kmeans.inertia_)

# Print the evaluation scores
print("Silhouette Coefficient:")
for k, score in silhouette_scores.items():
    print(f"K = {k}: {score:.4f}")
    
print("\nCalinski-Harabasz Index:")
for k, score in calinski_scores.items():
    print(f"K = {k}: {score:.4f}")
    
print("\nDavies-Bouldin Index:")
for k, score in davies_scores.items():
    print(f"K = {k}: {score:.4f}")

# Determine the best algorithm based on the evaluation scores
best_algo = max(silhouette_scores, calinski_scores, davies_scores, key=silhouette_scores.get)
best_k = best_algo

if best_algo == max(silhouette_scores, key=silhouette_scores.get):
    best_algo_name = "Silhouette Coefficient"
elif best_algo == max(calinski_scores, key=calinski_scores.get):
    best_algo_name = "Calinski-Harabasz Index"
else:
    best_algo_name = "Davies-Bouldin Index"

# Print the best algorithm and its K value
print(f"\nThe best algorithm based on the evaluation scores is {best_algo_name} with K = {best_k}")
```
